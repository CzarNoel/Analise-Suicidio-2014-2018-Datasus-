```python
#Importando Bibliotecas 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando os dados e chamando "enconding" para evitar Unicodeerror
data = pd.read_csv('data_suicidio.csv', encoding='latin9')
#Mostrando os primeiros dados do arquivo
data.head()
```

## Ocupação e Suicídio

##### A contagem de suicídios se mostra alta para os seguintes cargos de ocupação e situações de emprego/desemprego:

- Aposentados e Pensionistas
- Estudantes
- Donda de Casa
- Trabalhador Agropecuária em Geral
- Trabalhador Volante de Agricultura 
- Pedreiro
- Desemprego Crônico
- Empregado(a) doméstica nos Serviços Gerais

```python
# Código de ajuste do tamanho do gráfico
plt.figure(figsize=(15,15))
# Código de construção do gráfico
data['OCUP'].value_counts().plot.pie()
```

# O gráfico abaixo demonstra de maneira descendente os números de suicídio de acordo com a idade

De acordo com o mesmo, é possível observar que o maior número de casos de suicídio acontece aos **33** anos de idade.
Entretanto, a partir de um cálculo básico, a média de idade do total de número de suicídio, é **43** anos de idade.

```python
plt.figure(figsize = (20,5))
data['idade'].value_counts().plot.bar()
data['idade'].mean()
```
